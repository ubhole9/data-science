# Apriori Algorithm

#1.Support 10% and confidence 80%

```python
from mlxtend.frequent_patterns import apriori,association_rules
```

```python
#20% support
book_item=apriori(book,min_support=0.1,use_colnames=True)
book_item
```

```python
#80% confidence
book_rule1=association_rules(book_item,metric="lift",min_threshold=0.8)
book_rule1
```

```python
book_rule1.sort_values('lift',ascending = False)[0:20]
```

```python
# visualization of obtained rule
plt.scatter(book_rule1['support'],book_rule1['confidence'])
plt.xlabel('support')
plt.ylabel('confidence') 
plt.show()
```
